```python
#Files name
All_main_files=['fragments_data/gel24_main_fragments', #Gel24
                'fragments_data/gel21_fragments', #Gel 21
                'fragments_data/gel17_supp_fragments', #Gel 17]
                
            
All_N_initials=[[[937500, 734375,1062500,4005207,4825519],#Gel 24
                 [453120,765620,1421870,2674470,2638010], #Gel 21
                 [160937,201822,362238,365884,300259], #Gel 17]
                 
```

```python
import warnings
import pickle
warnings.filterwarnings("ignore", category=UserWarning)
from Fragment_Simulation_Functions import *
```

```python
#Uploading the data and initial values
data_labels=['0','CD4T', 'hESC-H9', 'HEK293T', 'RNH2-KO-T3-8','RNH2-KO-T3-17']

Total=10e6
n_reps=10
threshold_N=0.001

labels=['CD4T','hESC-H9', 'HEK293T','RNH2-KO-T3-8','RNH2-KO-T3-17']
```

# Run Simulation

```python
#Do the experiment for each label
for gel_index, main_file in enumerate(All_main_files):
    N_initial=All_N_initials[gel_index]
    for i,label in enumerate(labels):
        N_initial=N_initials[i]
        #Upload Data
        Treated, NonTreated=load_data(main_file+'_nontreated.csv',main_file+'_treated.csv',label,Total)
        Fragments=get_Fragments_DF(NonTreated,label)
        min_size=min(Fragments.index)

        #Run Binary Search on Simulations
        N,Fragments_try, total_cuts_try, new_lengths_try,medians_try,\
        Goal_Median, Initial_Median=binary_search_number_cuts(label, Treated, NonTreated, Fragments,
                                      Total,min_size,n_reps,threshold_N, N_initial)

        #Plot results
        print_plot_results(label, Treated, NonTreated, Fragments, Fragments_try, 
                           medians_try, total_cuts_try, Total,Goal_Median, Initial_Median,n_reps)

        #Compute statistics for ratios
        Total_DNA_bases=np.sum(np.array(Fragments.index)*np.array(Fragments[label]))
        All_cuts=np.mean([sum(F[label]) for F in Fragments_try])
        DNA_bases_New_Fragments=np.mean([sum(ll) for ll in new_lengths_try])
        New_cuts=np.mean(total_cuts_try)

        df = pd.DataFrame([Total_DNA_bases, DNA_bases_New_Fragments, All_cuts, New_cuts], 
                          columns=[label], index=['DNA bases', 'Avg. DNA bases in simulated new fragments',
                                                  'Number of fragments', 'Avg. No. of simulated cuts'])

        #Save statistics for ratios
        filename='outputs/Statistics_'+label+'_Total_'+format(float(Total),'.2')+'_reps_'+str(int(n_reps))+'.pkl'
        with open(filename, 'wb') as file:
            pickle.dump(df, file)

        #Get and save average distribution
        df = pd.concat(Fragments_try, axis=1).fillna(0)
        df['mean'] = df.apply(lambda row: np.mean(row[:-1]), axis=1)
        df=df[['mean']]
        df=df.reset_index()
        df.columns=['length',label]

        import pickle
        filename='outputs/Mean_Distribution_'+label+'_Total_'+format(float(Total),'.2')+'_reps_'+str(int(n_reps))+'.pkl'

        # Save the object to a pickle file
        with open(filename, 'wb') as file:
            pickle.dump(df, file)
```
